```julia
function RandUnit1(Range::Array{Float64,2}, m::Int)
    d = size(Range,1)
    Scales = Range[:,2].-Range[:,1]
    Rand = (rand(d,m).-1/2).*Scales
    for k = 1:m
        Rand[:,k]./=norm(Rand[:,k])
    end
    return Rand
end
```

```julia
function RandUnit2(Range::Array{Float64,2}, m::Int)
    ## d = size(Range,1)
    ## Scales = Range[:,2].-Range[:,1]
    Rand = (rand(size(Range,1),m).-1/2).*(Range[:,2].-Range[:,1])
    for k = 1:m
        Rand[:,k]./=norm(Rand[:,k])
    end
    return Rand
end
```
