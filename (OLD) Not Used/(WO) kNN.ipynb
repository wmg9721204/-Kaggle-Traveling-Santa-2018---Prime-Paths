```julia
N = 10000
d = 2
Sample = rand(d,N);
k = 20
```

```julia
function kNN1(Sample::Array{Float64,2}, k::Int)
    Dist = zeros(N,N)
    for i=1:N
        for j=i+1:N
            Dist[i,j] = sum((Sample[:,i]-Sample[:,j]).^2)
            Dist[j,i] = Dist[i,j]
        end
    end
    Dist
    kNN_Dict = Dict(i=>Array{Int}(sortslices([Dist[i,:] collect(1:N)], dims = 1)[2:k+1,2]) for i=1:N)
end
```

```julia
kNN1(Sample, 5)
```

```julia
Pts_Orders = zeros(Int, size(Sample))
d = size(Sample, 1)
N = size(Sample, 2)
for i = 1:2
    Pts_Seq = sortslices([Sample[i,:] collect(1:N)], dims = 1)
    Orders = Array{Int,1}(sortslices([Pts_Seq[:,2] collect(1:N)], dims = 1)[:,2])
    Pts_Orders[i,:] = Orders
end
Pts_Orders
```

```julia
GraphRep = zeros(Int, N,N)
for j=1:N
    GraphRep[Pts_Orders[:,j][1], Pts_Orders[:,j][2]] = j
end
GraphRep
```

```julia
using Clustering
```

```julia
n = 5
R = kmeans(Sample, n);
```

```julia
A = assignments(R)
C = counts(R);
```

```julia
using Plots
plotly()
Dict_Cluster = Dict(i=>Int[] for i=1:n)
scatter(Sample[1,:], Sample[2,:], label = "Sample")
for i=1:n
    Cluster_i = findall(A.==i)
    Dict_Cluster[i] = Cluster_i
    scatter!(Sample[1, Cluster_i], Sample[2, Cluster_i], label = "Cluster $i")
end
scatter!()
```
